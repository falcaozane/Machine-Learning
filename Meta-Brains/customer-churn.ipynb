```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset  = pd.read_csv('Churn_Modelling.csv')
```

```python
dataset
```

data cleaning spliting dataset into dependent and independent variables

```python
X = dataset.iloc[:,3:13].values
Y = dataset.iloc[:,13].values
```

```python
print(X)
```

```python
print(Y)
```

Label encoding using scikit for geography(hot encoding) and gender(0 or 1)

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label_X_2 = LabelEncoder()
X[:,2] = label_X_2.fit_transform(X[:,2])
```

```python
X
```

```python
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('ohe',OneHotEncoder(),[1])], remainder='passthrough')
X = np.array(ct.fit_transform(X), dtype = np.str_)
X = X[:,1:]
```

```python
print(X)
```

Spliting data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
pd.DataFrame(X_train)
```

Building the A-NN

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu', input_dim=11)) # layer 1
```

```python
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu')) # layer 2
```

```python
classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid')) # output layer
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.fit(X_train,Y_train,batch_size=10,epochs=100)
```

```python
Y_pred = classifier.predict(X_test)
Y_pred = (Y_pred>0.5)
```

```python
print(Y_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(Y_test,Y_pred)
```

```python
cm
```

```python
accuracy_score(Y_test,Y_pred)
```
