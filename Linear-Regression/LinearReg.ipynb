```python
import pandas as pd 

data = pd.read_csv("train.csv")

data.head()
```

```python
X = data['x']
Y = data['y']

X = X.tolist()
Y = Y.tolist()
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, Y, s = 6, color = '#88c999')
plt.grid()
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
#alpha - learning rate
def gradient_descent(X, Y, w, b, alpha):
 
    dl_dw = 0.0
    dl_db = 0.0
    N = len(X)

    for i in range(N):
        dl_dw += -1*X[i] * (Y[i] - (w*X[i] + b))
        dl_db += -1*(Y[i] - (w*X[i] + b))

    w = w - (1/float(N)) * dl_dw * alpha
    b = b - (1/float(N)) * dl_db * alpha

    return w, b
```

```python
def cost_function (X, Y, w, b):

    N = len(X)
    total_error = 0.0
    for i in range(N):
        total_error += (Y[i] - (w*X[i] - b))**2

    return total_error / (2*float(N))
```

```python
def train(X, Y, w, b, alpha, n_iter):

    for i in range(n_iter):
        w, b = gradient_descent(X, Y, w, b, alpha)

        if i % 400 == 0:
            print ("iteration:", i, "cost: ", cost_function(X, Y, w, b))


    return w, b
```

```python
def predict(x, w, b):
	return x*w + b
```

```python
w, b = train(X, Y, 0.0, 0.0, 0.0001, 7000)
x_new = 50.0
y_new = predict(x_new, w, b)
print(y_new)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Load the data
data = pd.read_csv("train.csv")

# Display the first few rows of the data
data.head()
```

```python
# Extract the features and target
X = data['x'].values.reshape(-1, 1)
Y = data['y'].values
```

```python
# Plot the data
plt.scatter(X, Y, s=6, color='#88c999')
plt.grid()
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```

```python
# Create a linear regression model
model = LinearRegression()
```

```python
# Train the model
model.fit(X, Y)
```

```python
# Print the coefficients
w = model.coef_[0]
b = model.intercept_
print(f"Slope (w): {w}")
print(f"Intercept (b): {b}")
```

```python
# Predict a new value
x_new = 50.0
y_new = model.predict(np.array([[x_new]]))
print(f"Predicted y for x={x_new}: {y_new[0]}")
```

```python
# Plot the regression line
plt.scatter(X, Y, s=5, color='#88c999')
plt.plot(X, model.predict(X), color='red')
plt.grid()
plt.xlabel("x values")
plt.ylabel("y values")
plt.show()
```
